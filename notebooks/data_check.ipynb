```python
import torch
```

```python
binary_data = "../src/arginator_protein_classifier/.data/processed_data_binary.pt"
multiclass_data = "../src/arginator_protein_classifier/.data/processed_data_multiclass.pt"
```

```python
print(type(binary_data))
print(type(multiclass_data))
```

```python
binary_data = torch.load(binary_data)
multiclass_data = torch.load(multiclass_data)
```

```python
print(type(binary_data))
print(type(multiclass_data))
```

```python
binary_data
```

```python
binary_embeddings, binary_labels = zip(*binary_data)
multi_embeddings, multi_labels = zip(*multiclass_data)
print(type(binary_labels), type(multi_labels))
```

```python
binary_labels_tensor = torch.tensor(binary_labels)
multi_labels_tensor = torch.tensor(multi_labels)
print(type(binary_labels_tensor))
```

```python
b_labels, freqs = torch.unique(binary_labels_tensor, return_counts=True)
m_labels, m_freqs = torch.unique(multi_labels_tensor, return_counts=True)
```

```python
for val, count in zip(m_labels, m_freqs):
    print(f"  Class {val.item()}: {count.item()} samples")
```
